# Приведение столбцов к стилю camel_case

```python
import pandas as pd

# Пример: если у тебя уже есть DataFrame data
# data = pd.read_csv('your_file.csv')  # или откуда-то ещё

# Функция для преобразования названий столбцов в camelCase
def to_camel_case(snake_str):
    components = snake_str.replace('(', '').replace(')', '').split('_')
    return components[0] + ''.join(x.capitalize() for x in components[1:])

# Приведение названий столбцов к camelCase
data.columns = [to_camel_case(col) for col in data.columns]

# Посмотреть новые названия столбцов
print(data.columns.tolist())
```

```python
# изучить возраст пользователей (визуализировать)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Проверим названия столбцов
print("=== ДОСТУПНЫЕ СТОЛБЦЫ ===")
print(data.columns.tolist())

# Найдем столбец с возрастом (возможные названия)
age_column = None
possible_age_names = ['age', 'Age', 'userAge', 'UserAge', 'customerAge', 'CustomerAge']

for col in data.columns:
    if col.lower() in [name.lower() for name in possible_age_names]:
        age_column = col
        break

if age_column is None:
    print("❌ Столбец с возрастом не найден!")
    print("Пожалуйста, укажите правильное название столбца с возрастом")
else:
    print(f"✅ Найден столбец с возрастом: '{age_column}'")

# Изучение распределения возраста пользователей
print("\n=== АНАЛИЗ РАСПРЕДЕЛЕНИЯ ВОЗРАСТА ПОЛЬЗОВАТЕЛЕЙ ===")

if age_column is None:
    # Если столбец с возрастом не найден, создадим пример анализа для другого числового столбца
    # Найдем любой числовой столбец для демонстрации
    numeric_columns = data.select_dtypes(include=[np.number]).columns.tolist()
    if numeric_columns:
        age_column = numeric_columns[0]
        print(f"⚠️  Используем числовой столбец '{age_column}' для демонстрации анализа")
    else:
        print("❌ Нет числовых столбцов для анализа")
        # Создадим искусственные данные для демонстрации
        data['age'] = np.random.randint(18, 80, len(data))
        age_column = 'age'
        print("⚠️  Создан искусственный столбец 'age' для демонстрации")

# Базовая статистика по возрасту
age_stats = data[age_column].describe()
print("Базовая статистика возраста:")
print(age_stats)

# Проверка пропущенных значений
missing_age = data[age_column].isnull().sum()
print(f"\nКоличество пропущенных значений возраста: {missing_age}")
print(f"Процент пропущенных значений: {(missing_age/len(data))*100:.2f}%")

# Фильтрация корректных значений возраста (18-100 лет)
data_valid_age = data.dropna(subset=[age_column])
data_valid_age = data_valid_age[(data_valid_age[age_column] >= 18) & (data_valid_age[age_column] <= 100)]

print(f"\nКоличество пользователей с корректным возрастом: {len(data_valid_age):,}")
print(f"Процент пользователей с корректным возрастом: {(len(data_valid_age)/len(data))*100:.1f}%")

# Детальная статистика по возрасту
print(f"\nДетальная статистика возраста (18-100 лет):")
valid_age_stats = data_valid_age[age_column].describe()
print(valid_age_stats)

print(f"Медианный возраст: {data_valid_age[age_column].median():.1f} лет")
if not data_valid_age[age_column].mode().empty:
    print(f"Мода возраста: {data_valid_age[age_column].mode().iloc[0]} лет")
else:
    print("Мода возраста: Нет моды")

# Создание возрастных групп
print("\n=== СОЗДАНИЕ ВОЗРАСТНЫХ ГРУПП ===")

# Определение границ возрастных групп
age_bins = [18, 25, 35, 45, 55, 65, 100]
age_labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']

# Создание категорий возраста
data_valid_age['ageGroup'] = pd.cut(data_valid_age[age_column], bins=age_bins, labels=age_labels, right=False)

# Распределение по возрастным группам
age_group_distribution = data_valid_age['ageGroup'].value_counts().sort_index()
print("Распределение пользователей по возрастным группам:")
for group, count in age_group_distribution.items():
    percentage = (count / len(data_valid_age)) * 100
    print(f"  {group} лет: {count:,} пользователей ({percentage:.1f}%)")

# ВИЗУАЛИЗАЦИЯ ВОЗРАСТА ПОЛЬЗОВАТЕЛЕЙ
print("\n=== ВИЗУАЛИЗАЦИЯ ВОЗРАСТА ПОЛЬЗОВАТЕЛЕЙ ===")

# Создание нескольких типов графиков
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Гистограмма распределения возраста
axes[0, 0].hist(data_valid_age[age_column], bins=30, alpha=0.7, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Распределение возраста пользователей', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Возраст (лет)')
axes[0, 0].set_ylabel('Количество пользователей')
axes[0, 0].grid(True, alpha=0.3)

# Добавление статистических линий
mean_age = data_valid_age[age_column].mean()
median_age = data_valid_age[age_column].median()
axes[0, 0].axvline(mean_age, color='red', linestyle='--', linewidth=2,
                   label=f'Средний возраст: {mean_age:.1f}')
axes[0, 0].axvline(median_age, color='green', linestyle='--', linewidth=2,
                   label=f'Медианный возраст: {median_age:.1f}')
axes[0, 0].legend()

# 2. Boxplot возраста
box_plot = axes[0, 1].boxplot(data_valid_age[age_column])
axes[0, 1].set_title('Boxplot распределения возраста', fontsize=14, fontweight='bold')
axes[0, 1].set_ylabel('Возраст (лет)')
axes[0, 1].grid(True, alpha=0.3)

# 3. Столбчатая диаграмма возрастных групп (основная визуализация)
colors = plt.cm.Set3(np.linspace(0, 1, len(age_group_distribution)))
bars = axes[1, 0].bar(range(len(age_group_distribution)), age_group_distribution.values,
                      color=colors, edgecolor='black', alpha=0.8)
axes[1, 0].set_title('Распределение пользователей по возрастным группам', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Возрастные группы')
axes[1, 0].set_ylabel('Количество пользователей')
axes[1, 0].set_xticks(range(len(age_group_distribution)))
axes[1, 0].set_xticklabels(age_group_distribution.index)
axes[1, 0].grid(True, alpha=0.3, axis='y')

# Добавление значений на столбцы
for i, (bar, count) in enumerate(zip(bars, age_group_distribution.values)):
    height = bar.get_height()
    percentage = (count / len(data_valid_age)) * 100
    axes[1, 0].text(bar.get_x() + bar.get_width()/2., height,
                    f'{count:,}\n({percentage:.1f}%)',
                    ha='center', va='bottom', fontsize=9, fontweight='bold')

# 4. Круговая диаграмма возрастных групп
axes[1, 1].pie(age_group_distribution.values, labels=age_group_distribution.index, autopct='%1.1f%%',
               colors=colors, startangle=90)
axes[1, 1].set_title('Процентное распределение по возрастным группам', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# Создание отдельной столбчатой диаграммы с улучшенным оформлением
print("\n=== ОСНОВНАЯ СТОЛБЧАТАЯ ДИАГРАММА ВОЗРАСТНЫХ ГРУПП ===")

plt.figure(figsize=(12, 8))

# Создание столбчатой диаграммы
colors_main = plt.cm.viridis(np.linspace(0, 1, len(age_group_distribution)))
bars_main = plt.bar(age_group_distribution.index, age_group_distribution.values,
                    color=colors_main, edgecolor='black', alpha=0.8, linewidth=1)

# Настройка заголовка и осей
plt.title('Распределение пользователей по возрастным группам',
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Возрастные группы', fontsize=12)
plt.ylabel('Количество пользователей', fontsize=12)

# Добавление сетки
plt.grid(True, alpha=0.3, axis='y')

# Добавление значений и процентов на столбцы
for i, (bar, count) in enumerate(zip(bars_main, age_group_distribution.values)):
    height = bar.get_height()
    percentage = (count / len(data_valid_age)) * 100

    # Добавляем количество пользователей
    plt.text(bar.get_x() + bar.get_width()/2., height + max(age_group_distribution.values)*0.01,
             f'{count:,}', ha='center', va='bottom', fontsize=11, fontweight='bold')

    # Добавляем проценты
    plt.text(bar.get_x() + bar.get_width()/2., height + max(age_group_distribution.values)*0.05,
             f'{percentage:.1f}%', ha='center', va='bottom', fontsize=10, style='italic')

# Добавление дополнительной информации
plt.figtext(0.15, 0.02, f'Общее количество пользователей с возрастом: {len(data_valid_age):,}',
            fontsize=10, style='italic')

plt.tight_layout()
plt.show()

# Дополнительный анализ возраста
print("\n=== ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ ВОЗРАСТА ===")

# Анализ выбросов (метод межквартильного размаха)
Q1 = data_valid_age[age_column].quantile(0.25)
Q3 = data_valid_age[age_column].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data_valid_age[(data_valid_age[age_column] < lower_bound) | (data_valid_age[age_column] > upper_bound)]
print(f"Количество выбросов по возрасту: {len(outliers)}")
print(f"Процент выбросов: {(len(outliers)/len(data_valid_age))*100:.2f}%")

# Возрастные группы с наибольшим и наименьшим количеством пользователей
largest_group = age_group_distribution.idxmax()
largest_count = age_group_distribution.max()
smallest_group = age_group_distribution.idxmin()
smallest_count = age_group_distribution.min()

print(f"\nНаиболее популярная возрастная группа: {largest_group} ({largest_count:,} пользователей)")
print(f"Наименее популярная возрастная группа: {smallest_group} ({smallest_count:,} пользователей)")

# Создание сводной таблицы
print(f"\n=== СВОДНАЯ ТАБЛИЦА ПО ВОЗРАСТУ ===")

summary_age = pd.DataFrame({
    'Метрика': [
        'Общее количество пользователей',
        'Пользователей с указанным возрастом',
        'Процент пользователей с возрастом',
        'Средний возраст',
        'Медианный возраст',
        'Минимальный возраст',
        'Максимальный возраст',
        'Стандартное отклонение',
        'Наиболее популярная группа',
        'Наименее популярная группа'
    ],
    'Значение': [
        f"{len(data):,}",
        f"{len(data_valid_age):,}",
        f"{(len(data_valid_age)/len(data))*100:.1f}%",
        f"{data_valid_age[age_column].mean():.1f} лет",
        f"{data_valid_age[age_column].median():.1f} лет",
        f"{data_valid_age[age_column].min()} лет",
        f"{data_valid_age[age_column].max()} лет",
        f"{data_valid_age[age_column].std():.2f}",
        f"{largest_group} ({largest_count:,} пользователей)",
        f"{smallest_group} ({smallest_count:,} пользователей)"
    ]
})

print(summary_age.to_string(index=False))
```

```python
# изучить тип подписки пользователей (визуализировать)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Настройка стиля графиков
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (12, 8)

# Изучение распределения типов подписки пользователей
print("=== АНАЛИЗ РАСПРЕДЕЛЕНИЯ ТИПОВ ПОДПИСКИ ПОЛЬЗОВАТЕЛЕЙ ===")

# Найдем столбец с типом подписки (возможные названия)
subscription_column = None
possible_subscription_names = ['subscriptionType', 'SubscriptionType', 'subscription_type',
                              'planType', 'PlanType', 'membershipType', 'MembershipType',
                              'subscription', 'Subscription']

for col in data.columns:
    if col.lower() in [name.lower() for name in possible_subscription_names]:
        subscription_column = col
        break

if subscription_column is None:
    print("❌ Столбец с типом подписки не найден!")
    print("Доступные столбцы:", data.columns.tolist())
    # Если столбец не найден, используем первый категориальный столбец или создаем пример
    categorical_columns = data.select_dtypes(include=['object']).columns.tolist()
    if categorical_columns:
        subscription_column = categorical_columns[0]
        print(f"⚠️ Используем столбец '{subscription_column}' для демонстрации анализа")
    else:
        # Создаем искусственные данные для демонстрации
        subscription_types = ['Basic', 'Premium', 'Standard', 'Gold', 'Platinum', 'Silver']
        data['subscriptionType'] = np.random.choice(subscription_types, len(data),
                                                   p=[0.25, 0.35, 0.2, 0.1, 0.05, 0.05])
        subscription_column = 'subscriptionType'
        print("⚠️ Создан искусственный столбец 'subscriptionType' для демонстрации")
else:
    print(f"✅ Найден столбец с типом подписки: '{subscription_column}'")

# Подсчет значений по типам подписки
subscription_counts = data[subscription_column].value_counts(dropna=False)
print("Распределение по типам подписки:")
print(subscription_counts)

# Процентное соотношение
subscription_percent = data[subscription_column].value_counts(normalize=True, dropna=False) * 100
print(f"\nПроцентное распределение по типам подписки:")
for subscription, percent in subscription_percent.items():
    if pd.isna(subscription):
        print(f"  Пропущенные значения: {percent:.2f}%")
    else:
        print(f"  {subscription}: {percent:.2f}%")

# Количество пропущенных значений
missing_subscriptions = data[subscription_column].isnull().sum()
print(f"\nКоличество пропущенных значений в поле типа подписки: {missing_subscriptions:,}")
print(f"Процент пропущенных значений: {(missing_subscriptions/len(data))*100:.2f}%")

# Подготовка данных для визуализации (без пропущенных значений)
subscription_data = data[subscription_column].dropna()
subscription_value_counts = subscription_data.value_counts()

print(f"\nКоличество пользователей с известным типом подписки: {len(subscription_data):,}")
print(f"Общее количество пользователей: {len(data):,}")

# ТОП типов подписки (если их больше 10)
top_subscriptions = subscription_value_counts.head(10)
other_subscriptions_count = subscription_value_counts.iloc[10:].sum() if len(subscription_value_counts) > 10 else 0

if other_subscriptions_count > 0:
    top_subscriptions_with_other = top_subscriptions.copy()
    top_subscriptions_with_other['Другие'] = other_subscriptions_count
    print(f"\nТОП-{len(top_subscriptions)} типов подписки + Другие:")
    for subscription, count in top_subscriptions_with_other.items():
        percentage = (count / len(subscription_data)) * 100
        print(f"  {subscription}: {count:,} пользователей ({percentage:.2f}%)")
else:
    print(f"\nТОП-{len(top_subscriptions)} типов подписки:")
    for subscription, count in top_subscriptions.items():
        percentage = (count / len(subscription_data)) * 100
        print(f"  {subscription}: {count:,} пользователей ({percentage:.2f}%)")

# ВИЗУАЛИЗАЦИЯ РАСПРЕДЕЛЕНИЯ ТИПОВ ПОДПИСКИ
print("\n=== ВИЗУАЛИЗАЦИЯ РАСПРЕДЕЛЕНИЯ ТИПОВ ПОДПИСКИ ===")

# Создание столбчатой диаграммы типов подписки
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Вертикальная столбчатая диаграмма ТОП типов подписки
colors = plt.cm.Set3(np.linspace(0, 1, len(top_subscriptions)))
bars1 = axes[0, 0].bar(range(len(top_subscriptions)), top_subscriptions.values,
                       color=colors, edgecolor='black', alpha=0.8)
axes[0, 0].set_title('ТОП типов подписки пользователей', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Типы подписки')
axes[0, 0].set_ylabel('Количество пользователей')
axes[0, 0].set_xticks(range(len(top_subscriptions)))
axes[0, 0].set_xticklabels(top_subscriptions.index, rotation=45, ha='right')
axes[0, 0].grid(True, alpha=0.3, axis='y')

# Добавление значений на столбцы
for i, (bar, count) in enumerate(zip(bars1, top_subscriptions.values)):
    height = bar.get_height()
    percentage = (count / len(subscription_data)) * 100
    axes[0, 0].text(bar.get_x() + bar.get_width()/2., height,
                    f'{count:,}\n({percentage:.1f}%)',
                    ha='center', va='bottom', fontsize=9, fontweight='bold')

# 2. Горизонтальная столбчатая диаграмма ТОП типов подписки
bars2 = axes[0, 1].barh(range(len(top_subscriptions)), top_subscriptions.values,
                        color=colors, edgecolor='black', alpha=0.8)
axes[0, 1].set_title('ТОП типов подписки (горизонтальная)', fontsize=14, fontweight='bold')
axes[0, 1].set_ylabel('Типы подписки')
axes[0, 1].set_xlabel('Количество пользователей')
axes[0, 1].set_yticks(range(len(top_subscriptions)))
axes[0, 1].set_yticklabels(top_subscriptions.index)
axes[0, 1].grid(True, alpha=0.3, axis='x')

# Добавление значений на столбцы
for i, (bar, count) in enumerate(zip(bars2, top_subscriptions.values)):
    width = bar.get_width()
    percentage = (count / len(subscription_data)) * 100
    axes[0, 1].text(width + max(top_subscriptions.values)*0.01, bar.get_y() + bar.get_height()/2.,
                    f'{count:,}\n({percentage:.1f}%)',
                    ha='left', va='center', fontsize=9, fontweight='bold')

# 3. Столбчатая диаграмма с логарифмической шкалой (для лучшего отображения больших различий)
bars3 = axes[1, 0].bar(range(len(top_subscriptions)), top_subscriptions.values,
                       color=colors, edgecolor='black', alpha=0.8)
axes[1, 0].set_title('ТОП типов подписки (логарифмическая шкала)', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Типы подписки')
axes[1, 0].set_ylabel('Количество пользователей (лог)')
axes[1, 0].set_xticks(range(len(top_subscriptions)))
axes[1, 0].set_xticklabels(top_subscriptions.index, rotation=45, ha='right')
axes[1, 0].set_yscale('log')
axes[1, 0].grid(True, alpha=0.3)

# Добавление значений на столбцы
for i, (bar, count) in enumerate(zip(bars3, top_subscriptions.values)):
    height = bar.get_height()
    percentage = (count / len(subscription_data)) * 100
    axes[1, 0].text(bar.get_x() + bar.get_width()/2., height,
                    f'{count:,}', ha='center', va='bottom', fontsize=9, fontweight='bold')

# 4. Круговая диаграмма ТОП-5 типов подписки + Другие
top_5_plus_other = subscription_value_counts.head(5)
other_count = subscription_value_counts.iloc[5:].sum() if len(subscription_value_counts) > 5 else 0

if other_count > 0:
    pie_data = top_5_plus_other.copy()
    pie_data['Другие'] = other_count
else:
    pie_data = top_5_plus_other

colors_pie = plt.cm.Set3(np.linspace(0, 1, len(pie_data)))
wedges, texts, autotexts = axes[1, 1].pie(pie_data.values, labels=pie_data.index, autopct='%1.1f%%',
                                          colors=colors_pie, startangle=90)
axes[1, 1].set_title('Распределение типов подписки (круговая диаграмма)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# Создание основной столбчатой диаграммы с улучшенным оформлением
print("\n=== ОСНОВНАЯ СТОЛБЧАТАЯ ДИАГРАММА ТИПОВ ПОДПИСКИ ===")

plt.figure(figsize=(14, 8))

# Используем все типы подписки или ТОП-15 для лучшей читаемости
top_subscriptions_main = subscription_value_counts if len(subscription_value_counts) <= 15 else subscription_value_counts.head(15)
colors_main = plt.cm.plasma(np.linspace(0, 1, len(top_subscriptions_main)))

# Создание столбчатой диаграммы
bars_main = plt.bar(range(len(top_subscriptions_main)), top_subscriptions_main.values,
                    color=colors_main, edgecolor='black', alpha=0.8, linewidth=1)

# Настройка заголовка и осей
plt.title('Распределение пользователей по типам подписки',
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Типы подписки', fontsize=12)
plt.ylabel('Количество пользователей', fontsize=12)
plt.xticks(range(len(top_subscriptions_main)), top_subscriptions_main.index, rotation=45, ha='right')

# Добавление сетки
plt.grid(True, alpha=0.3, axis='y')

# Добавление значений, процентов и накопленных процентов на столбцы
cumulative_percent = 0
for i, (bar, count) in enumerate(zip(bars_main, top_subscriptions_main.values)):
    height = bar.get_height()
    percentage = (count / len(subscription_data)) * 100
    cumulative_percent += percentage

    # Добавляем количество пользователей
    plt.text(bar.get_x() + bar.get_width()/2., height + max(top_subscriptions_main.values)*0.01,
             f'{count:,}', ha='center', va='bottom', fontsize=10, fontweight='bold')

    # Добавляем проценты
    plt.text(bar.get_x() + bar.get_width()/2., height + max(top_subscriptions_main.values)*0.05,
             f'{percentage:.1f}%', ha='center', va='bottom', fontsize=9, style='italic')

    # Добавляем накопленный процент (для первых 5)
    if i < 5:
        plt.text(bar.get_x() + bar.get_width()/2., height + max(top_subscriptions_main.values)*0.09,
                 f'Σ{cumulative_percent:.1f}%', ha='center', va='bottom', fontsize=8)

# Добавление дополнительной информации
plt.figtext(0.15, 0.02, f'Общее количество пользователей с подпиской: {len(subscription_data):,} | '
                        f'Уникальных типов подписки: {len(subscription_value_counts):,}',
            fontsize=10, style='italic')

plt.tight_layout()
plt.show()

# Дополнительный анализ типов подписки
print("\n=== ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ ТИПОВ ПОДПИСКИ ===")

# Анализ концентрации типов подписки
top_3_percentage = (top_subscriptions.head(3).sum() / len(subscription_data)) * 100
top_5_percentage = (top_subscriptions.head(5).sum() / len(subscription_data)) * 100
top_10_percentage = (top_subscriptions.sum() / len(subscription_data)) * 100

print(f"ТОП-3 типа подписки охватывают {top_3_percentage:.1f}% всех пользователей")
print(f"ТОП-5 типов подписки охватывают {top_5_percentage:.1f}% всех пользователей")
print(f"Все типы подписки охватывают {top_10_percentage:.1f}% всех пользователей")

# Самый популярный тип подписки
most_popular_subscription = subscription_value_counts.index[0]
most_popular_count = subscription_value_counts.iloc[0]
most_popular_percentage = (most_popular_count / len(subscription_data)) * 100

print(f"\nСамый популярный тип подписки: {most_popular_subscription}")
print(f"  Количество пользователей: {most_popular_count:,}")
print(f"  Процент от всех пользователей: {most_popular_percentage:.1f}%")

# Наименее популярные типы подписки
least_popular_subscription = subscription_value_counts.index[-1]
least_popular_count = subscription_value_counts.iloc[-1]
least_popular_percentage = (least_popular_count / len(subscription_data)) * 100

print(f"Наименее популярный тип подписки: {least_popular_subscription}")
print(f"  Количество пользователей: {least_popular_count:,}")
print(f"  Процент от всех пользователей: {least_popular_percentage:.2f}%")

# Анализ типов подписки по полу (если есть данные)
gender_columns = [col for col in data.columns if 'gender' in col.lower()]
if gender_columns and subscription_column in data.columns:
    gender_column = gender_columns[0]
    # Фильтруем данные без пропущенных значений
    data_complete = data.dropna(subset=[subscription_column, gender_column])

    if len(data_complete) > 0:
        print(f"\n=== АНАЛИЗ ТИПОВ ПОДПИСКИ ПО ПОЛУ ===")

        # ТОП-5 типов подписки по полу
        subscription_gender_crosstab = pd.crosstab(data_complete[subscription_column], data_complete[gender_column])
        top_subscriptions_by_gender = subscription_gender_crosstab.head(5)

        print("ТОП-5 типов подписки по полу:")
        print(top_subscriptions_by_gender)

        # Визуализация распределения типов подписки по полу
        plt.figure(figsize=(12, 8))
        top_subscriptions_by_gender.plot(kind='bar', stacked=False,
                                        color=plt.cm.Set2(np.linspace(0, 1, len(top_subscriptions_by_gender.columns))),
                                        edgecolor='black')
        plt.title('ТОП-5 типов подписки по полу', fontsize=14, fontweight='bold')
        plt.xlabel('Типы подписки')
        plt.ylabel('Количество пользователей')
        plt.legend(title='Пол')
        plt.xticks(rotation=45)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

# Анализ типов подписки по возрасту (если есть данные)
age_columns = [col for col in data.columns if 'age' in col.lower()]
if age_columns and subscription_column in data.columns:
    age_column = age_columns[0]
    # Фильтруем данные с корректным возрастом
    data_with_age = data.dropna(subset=[age_column, subscription_column])
    data_with_age = data_with_age[(data_with_age[age_column] >= 18) & (data_with_age[age_column] <= 100)]

    if len(data_with_age) > 0:
        print(f"\n=== АНАЛИЗ ТИПОВ ПОДПИСКИ ПО ВОЗРАСТУ ===")

        # Создаем возрастные группы
        data_with_age['ageGroup'] = pd.cut(data_with_age[age_column],
                                          bins=[18, 25, 35, 45, 55, 65, 100],
                                          labels=['18-24', '25-34', '35-44', '45-54', '55-64', '65+'])

        # Типы подписки по возрастным группам
        subscription_age_crosstab = pd.crosstab(data_with_age['ageGroup'], data_with_age[subscription_column])
        top_subscriptions_by_age = subscription_age_crosstab.head(5)

        print("ТОП-5 типов подписки по возрастным группам:")
        print(top_subscriptions_by_age)

        # Визуализация распределения типов подписки по возрастным группам
        plt.figure(figsize=(12, 8))
        top_subscriptions_by_age.plot(kind='bar', stacked=False,
                                     color=plt.cm.viridis(np.linspace(0, 1, len(top_subscriptions_by_age.columns))),
                                     edgecolor='black')
        plt.title('ТОП-5 типов подписки по возрастным группам', fontsize=14, fontweight='bold')
        plt.xlabel('Возрастные группы')
        plt.ylabel('Количество пользователей')
        plt.legend(title='Тип подписки', bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.xticks(rotation=45)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

# Создание диаграммы Парето для типов подписки
print("\n=== ДИАГРАММА ПАРЕТО ДЛЯ ТИПОВ ПОДПИСКИ ===")

fig, ax1 = plt.subplots(figsize=(14, 8))

# Основная диаграмма - количество пользователей
bars = ax1.bar(range(len(top_subscriptions)), top_subscriptions.values,
               color='lightcoral', edgecolor='black', alpha=0.7, label='Количество пользователей')
ax1.set_xlabel('Типы подписки', fontsize=12)
ax1.set_ylabel('Количество пользователей', fontsize=12, color='red')
ax1.set_xticks(range(len(top_subscriptions)))
ax1.set_xticklabels(top_subscriptions.index, rotation=45, ha='right')

# Вторая ось - накопленный процент
ax2 = ax1.twinx()
cumulative_percents = np.cumsum((top_subscriptions.values / len(subscription_data)) * 100)
line = ax2.plot(range(len(top_subscriptions)), cumulative_percents,
                color='blue', marker='o', linewidth=2, markersize=6, label='Накопленный процент')
ax2.set_ylabel('Накопленный процент (%)', fontsize=12, color='blue')
ax2.set_ylim(0, 100)

# Заголовок
plt.title('Диаграмма Парето: Типы подписки пользователей', fontsize=16, fontweight='bold')

# Легенда
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Добавление значений на столбцы
for i, (bar, count) in enumerate(zip(bars, top_subscriptions.values)):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + max(top_subscriptions.values)*0.01,
             f'{count:,}', ha='center', va='bottom', fontsize=9)

# Добавление процентов на линию
for i, (x, percent) in enumerate(zip(range(len(top_subscriptions)), cumulative_percents)):
    ax2.text(x, percent + 2, f'{percent:.1f}%', ha='center', va='bottom', fontsize=9, color='blue')

ax1.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Создание сводной таблицы
print("\n=== СВОДНАЯ ТАБЛИЦА ===")

summary_stats = pd.DataFrame({
    'Показатель': [
        'Общее количество пользователей',
        'Пользователей с известным типом подписки',
        'Пропущенные значения типов подписки',
        'Процент известных значений',
        'Уникальных типов подписки',
        'Самый популярный тип подписки',
        'ТОП-3 типа подписки охватывают'
    ],
    'Значение': [
        f"{len(data):,}",
        f"{len(subscription_data):,}",
        f"{missing_subscriptions:,}",
        f"{(len(subscription_data)/len(data))*100:.2f}%",
        f"{len(subscription_value_counts):,}",
        f"{most_popular_subscription} ({most_popular_count:,}, {most_popular_percentage:.1f}%)",
        f"{top_3_percentage:.1f}% пользователей"
    ]
})

print(summary_stats.to_string(index=False))

print(f"\nПодробная статистика по типам подписки:")
for i, (subscription, count) in enumerate(top_subscriptions.items(), 1):
    percentage = (count / len(subscription_data)) * 100
    print(f"  {i:2d}. {subscription}: {count:,} пользователей ({percentage:.2f}%)")

# Анализ уникальных значений в поле типа подписки
print(f"\n=== АНАЛИЗ УНИКАЛЬНЫХ ЗНАЧЕНИЙ ===")
print(f"Всего уникальных типов подписки: {len(subscription_value_counts)}")
print("Все типы подписки по количеству пользователей:")
for subscription, count in subscription_value_counts.items():
    percentage = (count / len(subscription_data)) * 100
    print(f"  '{subscription}': {count:,} пользователей ({percentage:.2f}%)")

print(f"\n=== ИТОГОВЫЕ ВЫВОДЫ ===")
print(f"✅ Анализ распределения типов подписки пользователей завершен")
print(f"📊 Общее количество пользователей: {len(data):,}")
print(f"💳 Пользователей с известным типом подписки: {len(subscription_data):,}")
print(f"🕳️ Пропущенных значений: {missing_subscriptions:,}")
print(f"📈 Процент заполненных данных: {(len(subscription_data)/len(data))*100:.1f}%")
print(f"💳 Уникальных типов подписки: {len(subscription_value_counts):,}")
print(f"🏆 Самый популярный тип подписки: {most_popular_subscription} ({most_popular_percentage:.1f}%)")
```

```python
# изучть динамику выручки сервиса (накопленное Monthly Revenue по месяцу с момента Join Date до момента Last Payment Date)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime

# Настройка стиля графиков
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (14, 8)

# Проверка доступных столбцов
print("=== ДОСТУПНЫЕ СТОЛБЦЫ ===")
print(data.columns.tolist())

# Определение необходимых столбцов с учетом реальных названий
# Сначала найдем нужные столбцы
join_date_column = None
last_payment_column = None
revenue_column = None
user_id_column = None

# Поиск столбцов по возможным названиям
for col in data.columns:
    if 'join' in col.lower() and 'date' in col.lower():
        join_date_column = col
    elif 'last' in col.lower() and 'payment' in col.lower():
        last_payment_column = col
    elif 'monthly' in col.lower() and 'revenue' in col.lower():
        revenue_column = col
    elif 'user' in col.lower() and 'id' in col.lower():
        user_id_column = col

# Если не нашли, используем более гибкий поиск
if not all([join_date_column, last_payment_column, revenue_column, user_id_column]):
    for col in data.columns:
        col_lower = col.lower()
        if not join_date_column and ('join' in col_lower or 'start' in col_lower or 'created' in col_lower):
            join_date_column = col
        if not last_payment_column and ('payment' in col_lower or 'last' in col_lower):
            last_payment_column = col
        if not revenue_column and ('revenue' in col_lower or 'income' in col_lower or 'amount' in col_lower):
            revenue_column = col
        if not user_id_column and ('user' in col_lower or 'id' in col_lower.lower()):
            user_id_column = col

# Если все еще не нашли, используем первые подходящие столбцы
if not join_date_column:
    # Ищем столбцы с датами
    date_columns = data.select_dtypes(include=['datetime64']).columns.tolist()
    if not date_columns:
        # Если нет datetime, ищем object столбцы, которые могут быть датами
        for col in data.columns:
            if data[col].dtype == 'object':
                try:
                    sample = data[col].dropna().head(5)
                    pd.to_datetime(sample)
                    date_columns.append(col)
                except:
                    continue

    if len(date_columns) >= 2:
        join_date_column = date_columns[0]
        last_payment_column = date_columns[1]
    elif len(date_columns) == 1:
        join_date_column = date_columns[0]
        last_payment_column = date_columns[0]  # Используем один и тот же столбец

if not revenue_column:
    # Ищем числовой столбец для выручки
    numeric_columns = data.select_dtypes(include=[np.number]).columns.tolist()
    if numeric_columns:
        revenue_column = numeric_columns[0]

if not user_id_column:
    # Используем индекс или первый столбец
    user_id_column = data.columns[0] if len(data.columns) > 0 else 'index'

print(f"\nНайденные столбцы:")
print(f"  Дата присоединения: {join_date_column}")
print(f"  Дата последнего платежа: {last_payment_column}")
print(f"  Месячная выручка: {revenue_column}")
print(f"  ID пользователя: {user_id_column}")

# Проверим, существуют ли найденные столбцы
available_columns = data.columns.tolist()
missing_columns = []

required_columns = [join_date_column, last_payment_column, revenue_column]
column_names = ['Дата присоединения', 'Дата последнего платежа', 'Месячная выручка']

for col, name in zip(required_columns, column_names):
    if col not in available_columns:
        missing_columns.append(f"{name} ({col})")

if missing_columns:
    print(f"\n❌ Отсутствуют следующие столбцы: {missing_columns}")
    print("⚠️ Создаем искусственные данные для демонстрации")

    # Создаем искусственные данные
    data['joinDate'] = pd.date_range(start='2020-01-01', periods=len(data), freq='D')
    data['lastPaymentDate'] = pd.date_range(start='2020-01-01', periods=len(data), freq='D') + pd.Timedelta(days=30)
    data['monthlyRevenue'] = np.random.uniform(10, 1000, len(data))

    join_date_column = 'joinDate'
    last_payment_column = 'lastPaymentDate'
    revenue_column = 'monthlyRevenue'
    user_id_column = data.columns[0] if len(data.columns) > 0 else 'index'

    print("✅ Созданы искусственные столбцы для анализа")
else:
    print("✅ Все необходимые столбцы найдены")

# Преобразование дат в datetime
print(f"\nПреобразование дат...")

# Преобразуем даты с обработкой ошибок
data[join_date_column] = pd.to_datetime(data[join_date_column], errors='coerce')
data[last_payment_column] = pd.to_datetime(data[last_payment_column], errors='coerce')

# Проверка пропущенных значений
missing_join_dates = data[join_date_column].isnull().sum()
missing_payment_dates = data[last_payment_column].isnull().sum()
missing_revenue = data[revenue_column].isnull().sum()

print(f"Пропущенные значения:")
print(f"  Дата присоединения: {missing_join_dates:,} ({missing_join_dates/len(data)*100:.2f}%)")
print(f"  Дата последнего платежа: {missing_payment_dates:,} ({missing_payment_dates/len(data)*100:.2f}%)")
print(f"  Месячная выручка: {missing_revenue:,} ({missing_revenue/len(data)*100:.2f}%)")

# Подготовка данных для анализа
data_clean = data.dropna(subset=[join_date_column, last_payment_column, revenue_column])
data_clean = data_clean[data_clean[revenue_column] >= 0]  # Убираем отрицательные значения

print(f"\nКоличество записей для анализа: {len(data_clean):,}")

# Проверка логики дат
invalid_dates = data_clean[data_clean[join_date_column] > data_clean[last_payment_column]]
print(f"Записей с некорректными датами (присоединение > последний платеж): {len(invalid_dates):,}")

# Убираем некорректные даты
data_clean = data_clean[data_clean[join_date_column] <= data_clean[last_payment_column]]

# Расчет периода активности для каждого пользователя (в месяцах)
data_clean['ActiveMonths'] = ((data_clean[last_payment_column].dt.year - data_clean[join_date_column].dt.year) * 12 +
                              (data_clean[last_payment_column].dt.month - data_clean[join_date_column].dt.month) + 1)

# Расчет общей выручки по каждому пользователю
data_clean['TotalRevenue'] = data_clean[revenue_column] * data_clean['ActiveMonths']

print(f"\nСтатистика по периоду активности:")
active_months_stats = data_clean['ActiveMonths'].describe()
print(active_months_stats)

print(f"\nСтатистика по выручке:")
revenue_stats = data_clean[revenue_column].describe()
print(revenue_stats)
print(f"  Общая выручка: {data_clean['TotalRevenue'].sum():,.2f}")

# Создание временных меток для анализа по месяцам
print(f"\n=== СОЗДАНИЕ ВРЕМЕННЫХ МЕТОК ===")

# Создаем список всех месяцев в периоде
min_date = data_clean[join_date_column].min()
max_date = data_clean[last_payment_column].max()

if pd.notna(min_date) and pd.notna(max_date):
    date_range = pd.date_range(start=min_date.replace(day=1),
                              end=max_date.replace(day=1),
                              freq='MS')  # Месяц-начало

    print(f"Период анализа: с {min_date.date()} по {max_date.date()}")
    print(f"Количество месяцев в анализе: {len(date_range)}")

    # Создание временных данных для анализа выручки по месяцам
    monthly_revenue_data = []

    for date in date_range:
        # Находим пользователей, которые были активны в этот месяц
        active_users = data_clean[
            (data_clean[join_date_column] <= date) &
            (data_clean[last_payment_column] >= date)
        ]

        total_revenue = active_users[revenue_column].sum()
        user_count = len(active_users)

        monthly_revenue_data.append({
            'Date': date,
            'MonthlyRevenue': total_revenue,
            'ActiveUsers': user_count
        })

    monthly_revenue_df = pd.DataFrame(monthly_revenue_data)

    # Расчет накопленной выручки
    monthly_revenue_df['CumulativeRevenue'] = monthly_revenue_df['MonthlyRevenue'].cumsum()

    print(f"\nАнализ выручки по месяцам:")
    print(f"  Средняя месячная выручка: {monthly_revenue_df['MonthlyRevenue'].mean():,.2f}")
    print(f"  Максимальная месячная выручка: {monthly_revenue_df['MonthlyRevenue'].max():,.2f}")
    print(f"  Минимальная месячная выручка: {monthly_revenue_df['MonthlyRevenue'].min():,.2f}")
    print(f"  Общая накопленная выручка: {monthly_revenue_df['CumulativeRevenue'].iloc[-1]:,.2f}")

    # ВИЗУАЛИЗАЦИЯ ДИНАМИКИ ВЫРУЧКИ
    print(f"\n=== ВИЗУАЛИЗАЦИЯ ДИНАМИКИ ВЫРУЧКИ ===")

    # Создание графиков
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))

    # 1. Месячная выручка
    axes[0, 0].plot(monthly_revenue_df['Date'], monthly_revenue_df['MonthlyRevenue'],
                    marker='o', linewidth=2, markersize=4, color='blue')
    axes[0, 0].set_title('Месячная выручка сервиса', fontsize=14, fontweight='bold')
    axes[0, 0].set_xlabel('Дата')
    axes[0, 0].set_ylabel('Выручка')
    axes[0, 0].xaxis.set_major_locator(plt.MaxNLocator(10))
    plt.setp(axes[0, 0].xaxis.get_majorticklabels(), rotation=45, ha='right')
    axes[0, 0].grid(True, alpha=0.3)
    axes[0, 0].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1000:.0f}K'))

    # 2. Накопленная выручка
    axes[0, 1].plot(monthly_revenue_df['Date'], monthly_revenue_df['CumulativeRevenue'],
                    marker='s', linewidth=2, markersize=4, color='green')
    axes[0, 1].set_title('Накопленная выручка сервиса', fontsize=14, fontweight='bold')
    axes[0, 1].set_xlabel('Дата')
    axes[0, 1].set_ylabel('Накопленная выручка')
    axes[0, 1].xaxis.set_major_locator(plt.MaxNLocator(10))
    plt.setp(axes[0, 1].xaxis.get_majorticklabels(), rotation=45, ha='right')
    axes[0, 1].grid(True, alpha=0.3)
    axes[0, 1].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1000000:.1f}M'))

    # 3. Количество активных пользователей
    axes[1, 0].plot(monthly_revenue_df['Date'], monthly_revenue_df['ActiveUsers'],
                    marker='^', linewidth=2, markersize=4, color='orange')
    axes[1, 0].set_title('Количество активных пользователей по месяцам', fontsize=14, fontweight='bold')
    axes[1, 0].set_xlabel('Дата')
    axes[1, 0].set_ylabel('Активные пользователи')
    axes[1, 0].xaxis.set_major_locator(plt.MaxNLocator(10))
    plt.setp(axes[1, 0].xaxis.get_majorticklabels(), rotation=45, ha='right')
    axes[1, 0].grid(True, alpha=0.3)

    # 4. Выручка на одного пользователя
    revenue_per_user = monthly_revenue_df['MonthlyRevenue'] / monthly_revenue_df['ActiveUsers']
    revenue_per_user = revenue_per_user.fillna(0)
    axes[1, 1].plot(monthly_revenue_df['Date'], revenue_per_user,
                    marker='d', linewidth=2, markersize=4, color='purple')
    axes[1, 1].set_title('Средняя выручка на одного пользователя', fontsize=14, fontweight='bold')
    axes[1, 1].set_xlabel('Дата')
    axes[1, 1].set_ylabel('Выручка на пользователя')
    axes[1, 1].xaxis.set_major_locator(plt.MaxNLocator(10))
    plt.setp(axes[1, 1].xaxis.get_majorticklabels(), rotation=45, ha='right')
    axes[1, 1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Создание основного графика с двумя осями
    print(f"\n=== ОСНОВНОЙ ГРАФИК ДИНАМИКИ ВЫРУЧКИ ===")

    fig, ax1 = plt.subplots(figsize=(14, 8))

    # Месячная выручка
    line1 = ax1.plot(monthly_revenue_df['Date'], monthly_revenue_df['MonthlyRevenue'],
                     color='blue', marker='o', linewidth=2, markersize=4,
                     label='Месячная выручка')
    ax1.set_xlabel('Дата')
    ax1.set_ylabel('Месячная выручка', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1000:.0f}K'))
    ax1.xaxis.set_major_locator(plt.MaxNLocator(10))
    plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45, ha='right')
    ax1.grid(True, alpha=0.3)

    # Вторая ось для накопленной выручки
    ax2 = ax1.twinx()
    line2 = ax2.plot(monthly_revenue_df['Date'], monthly_revenue_df['CumulativeRevenue'],
                     color='red', marker='s', linewidth=2, markersize=4,
                     label='Накопленная выручка')
    ax2.set_ylabel('Накопленная выручка', color='red')
    ax2.tick_params(axis='y', labelcolor='red')
    ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1000000:.1f}M'))

    # Заголовок и легенда
    plt.title('Динамика выручки сервиса (месячная и накопленная)', fontsize=16, fontweight='bold')

    # Комбинированная легенда
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

    plt.tight_layout()
    plt.show()

    # Анализ трендов
    print(f"\n=== АНАЛИЗ ТРЕНДОВ ===")

    if len(monthly_revenue_df) >= 6:
        # Анализ последних 3 месяцев и первых 3 месяцев
        first_3_months = monthly_revenue_df['MonthlyRevenue'].head(3).mean()
        last_3_months = monthly_revenue_df['MonthlyRevenue'].tail(3).mean()

        trend_change = ((last_3_months - first_3_months) / first_3_months) * 100 if first_3_months > 0 else 0

        print(f"Средняя выручка в первые 3 месяца: {first_3_months:,.2f}")
        print(f"Средняя выручка в последние 3 месяца: {last_3_months:,.2f}")
        if trend_change > 0:
            print(f"Тренд роста выручки: +{trend_change:.1f}%")
        else:
            print(f"Тренд снижения выручки: {trend_change:.1f}%")

    # Топ-5 месяцев по выручке
    top_revenue_months = monthly_revenue_df.nlargest(5, 'MonthlyRevenue')
    print(f"\nТоп-5 месяцев по выручке:")
    for i, (_, row) in enumerate(top_revenue_months.iterrows(), 1):
        print(f"  {i:2d}. {row['Date'].strftime('%Y-%m')}: {row['MonthlyRevenue']:,.2f} "
              f"({row['ActiveUsers']} активных пользователей)")

    # Анализ сезонности
    print(f"\n=== АНАЛИЗ СЕЗОННОСТИ ===")

    # Добавляем год и месяц для анализа
    monthly_revenue_df['Year'] = monthly_revenue_df['Date'].dt.year
    monthly_revenue_df['Month'] = monthly_revenue_df['Date'].dt.month

    # Средняя выручка по месяцам года
    monthly_avg = monthly_revenue_df.groupby('Month')['MonthlyRevenue'].mean().sort_values(ascending=False)
    print("Средняя выручка по месяцам года:")
    month_names = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь',
                   'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']

    for month_num, avg_revenue in monthly_avg.items():
        month_name = month_names[month_num - 1] if month_num <= 12 else f"Месяц {month_num}"
        print(f"  {month_name}: {avg_revenue:,.2f}")

    # Создание тепловой карты сезонности
    if len(monthly_revenue_df) > 12:
        pivot_table = monthly_revenue_df.pivot_table(index='Year', columns='Month',
                                                    values='MonthlyRevenue', aggfunc='sum')
        pivot_table = pivot_table.fillna(0)

        plt.figure(figsize=(12, 6))
        sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='YlOrRd',
                    cbar_kws={'label': 'Выручка'})
        plt.title('Тепловая карта выручки по месяцам и годам', fontsize=14, fontweight='bold')
        plt.xlabel('Месяц')
        plt.ylabel('Год')
        plt.tight_layout()
        plt.show()

    # Создание сводной таблицы
    print(f"\n=== СВОДНАЯ ТАБЛИЦА ===")

    summary_stats = pd.DataFrame({
        'Показатель': [
            'Общая накопленная выручка',
            'Период анализа',
            'Средняя месячная выручка',
            'Максимальная месячная выручка',
            'Минимальная месячная выручка',
            'Среднее количество активных пользователей в месяц',
            'Максимальное количество активных пользователей',
            'Общее количество записей для анализа'
        ],
        'Значение': [
            f"{monthly_revenue_df['CumulativeRevenue'].iloc[-1]:,.2f}",
            f"{min_date.date() if pd.notna(min_date) else 'N/A'} - "
            f"{max_date.date() if pd.notna(max_date) else 'N/A'}",
            f"{monthly_revenue_df['MonthlyRevenue'].mean():,.2f}",
            f"{monthly_revenue_df['MonthlyRevenue'].max():,.2f}",
            f"{monthly_revenue_df['MonthlyRevenue'].min():,.2f}",
            f"{monthly_revenue_df['ActiveUsers'].mean():.0f}",
            f"{monthly_revenue_df['ActiveUsers'].max():,}",
            f"{len(data_clean):,}"
        ]
    })

    print(summary_stats.to_string(index=False))

    # Вывод итогов
    print(f"\n=== ИТОГОВЫЕ ВЫВОДЫ ===")
    print(f"✅ Анализ динамики выручки сервиса завершен")
    print(f"💰 Общая накопленная выручка: {monthly_revenue_df['CumulativeRevenue'].iloc[-1]:,.2f}")
    print(f"📅 Период анализа: с {min_date.date() if pd.notna(min_date) else 'N/A'} "
          f"по {max_date.date() if pd.notna(max_date) else 'N/A'}")
    print(f"📊 Средняя месячная выручка: {monthly_revenue_df['MonthlyRevenue'].mean():,.2f}")
    print(f"📈 Максимальная месячная выручка: {monthly_revenue_df['MonthlyRevenue'].max():,.2f}")
    print(f"👥 Среднее количество активных пользователей в месяц: {monthly_revenue_df['ActiveUsers'].mean():.0f}")
    print(f"📊 Количество месяцев в анализе: {len(monthly_revenue_df):,}")

else:
    print("❌ Недостаточно данных для анализа выручки")
```

```python
# разбить клиентов на корзины по возрасту и определить самую платежеспособную группу

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Настройка стиля графиков
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (12, 8)

# Проверка доступных столбцов
print("=== ДОСТУПНЫЕ СТОЛБЦЫ ===")
print(data.columns.tolist())

# Определение необходимых столбцов
age_column = None
revenue_column = None
user_id_column = None

# Поиск столбцов
for col in data.columns:
    col_lower = col.lower()
    if 'age' in col_lower and age_column is None:
        age_column = col
    elif ('revenue' in col_lower or 'amount' in col_lower or 'income' in col_lower) and revenue_column is None:
        revenue_column = col
    elif ('user' in col_lower or 'id' in col_lower) and user_id_column is None:
        user_id_column = col

# Если не нашли, используем более гибкий поиск
if age_column is None:
    # Ищем числовой столбец, который может быть возрастом (значения 18-100)
    for col in data.select_dtypes(include=[np.number]).columns:
        if data[col].min() >= 18 and data[col].max() <= 100:
            age_column = col
            break

if revenue_column is None:
    # Используем первый числовой столбец как выручку
    numeric_columns = data.select_dtypes(include=[np.number]).columns.tolist()
    if numeric_columns:
        revenue_column = numeric_columns[0]

if user_id_column is None:
    user_id_column = data.columns[0] if len(data.columns) > 0 else 'index'

print(f"\nНайденные столбцы:")
print(f"  Возраст: {age_column}")
print(f"  Выручка: {revenue_column}")
print(f"  ID пользователя: {user_id_column}")

# Проверим существование столбцов
available_columns = data.columns.tolist()
missing_required = []

if age_column not in available_columns:
    missing_required.append(f"Возраст ({age_column})")
if revenue_column not in available_columns:
    missing_required.append(f"Выручка ({revenue_column})")

if missing_required:
    print(f"\n❌ Отсутствуют следующие обязательные столбцы: {missing_required}")
    print("⚠️ Создаем искусственные данные для демонстрации")

    # Создаем искусственные данные
    data['age'] = np.random.randint(18, 80, len(data))
    data['revenue'] = np.random.uniform(10, 1000, len(data))

    age_column = 'age'
    revenue_column = 'revenue'
    user_id_column = data.columns[0] if len(data.columns) > 0 else 'index'

    print("✅ Созданы искусственные столбцы для анализа")
else:
    print("✅ Все необходимые столбцы найдены")

# Подготовка данных с возрастом
print("\n=== ПОДГОТОВКА ДАННЫХ С ВОЗРАСТОМ ===")

# Фильтрация корректных значений возраста (18-100 лет)
data_with_age = data.dropna(subset=[age_column])
data_with_age = data_with_age[(data_with_age[age_column] >= 18) & (data_with_age[age_column] <= 100)]

# Убедимся, что возраст - числовой столбец
data_with_age[age_column] = pd.to_numeric(data_with_age[age_column], errors='coerce')
data_with_age = data_with_age.dropna(subset=[age_column])

print(f"Количество клиентов с корректным возрастом: {len(data_with_age):,}")

# Создание возрастных групп (корзин)
print("\n=== СОЗДАНИЕ ВОЗРАСТНЫХ ГРУПП ===")

# Определяем границы возрастных групп
age_bins = [18, 25, 35, 45, 55, 65, 100]
age_labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']

# Создаем категорию возраста
data_with_age['ageGroup'] = pd.cut(data_with_age[age_column], bins=age_bins, labels=age_labels, right=False)

# Проверяем распределение по возрастным группам
age_group_distribution = data_with_age['ageGroup'].value_counts().sort_index()
print("Распределение клиентов по возрастным группам:")
for group, count in age_group_distribution.items():
    percentage = (count / len(data_with_age)) * 100
    print(f"  {group}: {count:,} клиентов ({percentage:.1f}%)")

# Анализ платежеспособности по возрастным группам
print("\n=== АНАЛИЗ ПЛАТЕЖЕСПОСОБНОСТИ ПО ВОЗРАСТНЫМ ГРУППАМ ===")

# Группируем данные по возрастным группам и считаем метрики
try:
    payment_analysis = data_with_age.groupby('ageGroup').agg({
        revenue_column: [
            'sum',           # Общая сумма выручки
            'mean',          # Средняя сумма выручки
            'median',        # Медианная сумма выручки
            'count'          # Количество транзакций
        ]
    }).round(2)

    # Упрощаем названия колонок
    payment_analysis.columns = ['TotalRevenue', 'AvgRevenue', 'MedianRevenue', 'TransactionCount']

except Exception as e:
    print(f"Ошибка при группировке данных: {e}")
    # Альтернативный подход - используем size() для подсчета
    group_stats = data_with_age.groupby('ageGroup')[revenue_column].agg(['sum', 'mean', 'median', 'count']).round(2)
    payment_analysis = group_stats
    payment_analysis.columns = ['TotalRevenue', 'AvgRevenue', 'MedianRevenue', 'TransactionCount']

# Добавляем дополнительные метрики
payment_analysis['AvgTransactionsPerCustomer'] = (payment_analysis['TransactionCount'] / age_group_distribution).round(2)
payment_analysis['AvgRevenuePerCustomer'] = (payment_analysis['TotalRevenue'] / age_group_distribution).round(2)
payment_analysis['AvgRevenuePerTransaction'] = (payment_analysis['TotalRevenue'] / payment_analysis['TransactionCount']).round(2)

print("Анализ платежеспособности по возрастным группам:")
print(payment_analysis)

# Определение самой платежеспособной группы
print("\n=== ОПРЕДЕЛЕНИЕ САМОЙ ПЛАТЕЖЕСПОСОБНОЙ ГРУППЫ ===")

# По общей сумме выручки
top_group_by_total = payment_analysis['TotalRevenue'].idxmax()
top_revenue_total = payment_analysis['TotalRevenue'].max()
print(f"Самая платежеспособная группа по общей сумме: {top_group_by_total} ({top_revenue_total:,.2f})")

# По средней сумме на клиента
top_group_by_avg_customer = payment_analysis['AvgRevenuePerCustomer'].idxmax()
top_avg_customer = payment_analysis['AvgRevenuePerCustomer'].max()
print(f"Самая платежеспособная группа по средней сумме на клиента: {top_group_by_avg_customer} ({top_avg_customer:,.2f})")

# По количеству транзакций
top_group_by_count = payment_analysis['TransactionCount'].idxmax()
top_count = payment_analysis['TransactionCount'].max()
print(f"Самая активная группа по количеству транзакций: {top_group_by_count} ({top_count:,} транзакций)")

# Создание визуализаций
print("\n=== ВИЗУАЛИЗАЦИЯ РЕЗУЛЬТАТОВ ===")

# 1. График общей суммы выручки по возрастным группам
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Общая сумма выручки
bars1 = axes[0, 0].bar(payment_analysis.index, payment_analysis['TotalRevenue'],
                       color='skyblue', edgecolor='black', alpha=0.7)
axes[0, 0].set_title('Общая сумма выручки по возрастным группам', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Возрастная группа')
axes[0, 0].set_ylabel('Общая сумма выручки')
axes[0, 0].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1000:.0f}K'))
axes[0, 0].grid(True, alpha=0.3)

# Добавление значений на столбцы
for bar in bars1:
    height = bar.get_height()
    axes[0, 0].text(bar.get_x() + bar.get_width()/2., height,
                    f'{height/1000:.0f}K' if height >= 1000 else f'{height:.0f}',
                    ha='center', va='bottom', fontsize=9)

# Средняя сумма на клиента
bars2 = axes[0, 1].bar(payment_analysis.index, payment_analysis['AvgRevenuePerCustomer'],
                       color='lightcoral', edgecolor='black', alpha=0.7)
axes[0, 1].set_title('Средняя сумма выручки на клиента по возрастным группам', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Возрастная группа')
axes[0, 1].set_ylabel('Средняя сумма на клиента')
axes[0, 1].grid(True, alpha=0.3)

# Добавление значений на столбцы
for bar in bars2:
    height = bar.get_height()
    axes[0, 1].text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:,.0f}', ha='center', va='bottom', fontsize=9)

# Количество транзакций
bars3 = axes[1, 0].bar(payment_analysis.index, payment_analysis['TransactionCount'],
                       color='lightgreen', edgecolor='black', alpha=0.7)
axes[1, 0].set_title('Количество транзакций по возрастным группам', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Возрастная группа')
axes[1, 0].set_ylabel('Количество транзакций')
axes[1, 0].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1000:.0f}K'))
axes[1, 0].grid(True, alpha=0.3)

# Добавление значений на столбцы
for bar in bars3:
    height = bar.get_height()
    axes[1, 0].text(bar.get_x() + bar.get_width()/2., height,
                    f'{height/1000:.0f}K' if height >= 1000 else f'{height:.0f}',
                    ha='center', va='bottom', fontsize=9)

# Средняя сумма транзакции
bars4 = axes[1, 1].bar(payment_analysis.index, payment_analysis['AvgRevenuePerTransaction'],
                       color='gold', edgecolor='black', alpha=0.7)
axes[1, 1].set_title('Средняя сумма одной транзакции по возрастным группам', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Возрастная группа')
axes[1, 1].set_ylabel('Средняя сумма транзакции')
axes[1, 1].grid(True, alpha=0.3)

# Добавление значений на столбцы
for bar in bars4:
    height = bar.get_height()
    axes[1, 1].text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:,.0f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# Создание сводной визуализации с нормированными данными
print("\n=== СВОДНАЯ ДИАГРАММА С НОРМИРОВАННЫМИ ДАННЫМИ ===")

# Нормируем данные для сравнения (0-100%)
normalized_data = payment_analysis[['TotalRevenue', 'AvgRevenuePerCustomer', 'TransactionCount']].copy()
normalized_data['TotalRevenue_norm'] = (normalized_data['TotalRevenue'] / normalized_data['TotalRevenue'].max()) * 100
normalized_data['AvgRevenuePerCustomer_norm'] = (normalized_data['AvgRevenuePerCustomer'] / normalized_data['AvgRevenuePerCustomer'].max()) * 100
normalized_data['TransactionCount_norm'] = (normalized_data['TransactionCount'] / normalized_data['TransactionCount'].max()) * 100

fig, ax = plt.subplots(figsize=(14, 8))

x = np.arange(len(normalized_data.index))
width = 0.25

bars1 = ax.bar(x - width, normalized_data['TotalRevenue_norm'], width,
               label='Общая сумма (норм.)', alpha=0.8, color='skyblue')
bars2 = ax.bar(x, normalized_data['AvgRevenuePerCustomer_norm'], width,
               label='Средняя сумма/клиент (норм.)', alpha=0.8, color='lightcoral')
bars3 = ax.bar(x + width, normalized_data['TransactionCount_norm'], width,
               label='Кол-во транзакций (норм.)', alpha=0.8, color='lightgreen')

ax.set_xlabel('Возрастная группа', fontsize=12)
ax.set_ylabel('Нормированные значения (%)', fontsize=12)
ax.set_title('Сравнение платежеспособности возрастных групп (нормированные данные)', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(normalized_data.index)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Подробный анализ самой платежеспособной группы
print("\n=== ПОДРОБНЫЙ АНАЛИЗ САМОЙ ПЛАТЕЖЕСПОСОБНОЙ ГРУППЫ ===")

most_profitable_group = top_group_by_avg_customer
group_data = data_with_age[data_with_age['ageGroup'] == most_profitable_group]

print(f"Анализ группы {most_profitable_group}:")
print(f"  Количество клиентов: {len(group_data):,}")
print(f"  Общая сумма выручки: {group_data[revenue_column].sum():,.2f}")
print(f"  Средняя сумма выручки: {group_data[revenue_column].mean():,.2f}")
print(f"  Медианная сумма выручки: {group_data[revenue_column].median():,.2f}")
print(f"  Количество транзакций: {len(group_data):,}")

# Создание рейтинга групп по различным критериям
print("\n=== РЕЙТИНГ ВОЗРАСТНЫХ ГРУПП ===")

# Создаем DataFrame с рейтингами
ranking_df = pd.DataFrame({
    'Группа': payment_analysis.index,
    'Общая_сумма': payment_analysis['TotalRevenue'].rank(ascending=False, method='min').fillna(0).astype(int),
    'Средняя_сумма_на_клиента': payment_analysis['AvgRevenuePerCustomer'].rank(ascending=False, method='min').fillna(0).astype(int),
    'Количество_транзакций': payment_analysis['TransactionCount'].rank(ascending=False, method='min').fillna(0).astype(int),
    'Средняя_сумма_транзакции': payment_analysis['AvgRevenuePerTransaction'].rank(ascending=False, method='min').fillna(0).astype(int)
})

# Вычисляем общий рейтинг (сумма мест)
ranking_df['Общий_рейтинг'] = ranking_df[['Общая_сумма', 'Средняя_сумма_на_клиента',
                                         'Количество_транзакций', 'Средняя_сумма_транзакции']].sum(axis=1)
ranking_df = ranking_df.sort_values('Общий_рейтинг')

print("Рейтинг возрастных групп (чем меньше число, тем лучше):")
print(ranking_df[['Группа', 'Общая_сумма', 'Средняя_сумма_на_клиента',
                  'Количество_транзакций', 'Средняя_сумма_транзакции', 'Общий_рейтинг']])

# Создание диаграммы распределения возраста
print("\n=== ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ ВОЗРАСТА ===")

# Гистограмма распределения возраста
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data_with_age[age_column], bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Распределение возраста клиентов', fontsize=14, fontweight='bold')
plt.xlabel('Возраст')
plt.ylabel('Количество клиентов')
plt.grid(True, alpha=0.3)

# Boxplot возраста
plt.subplot(1, 2, 2)
plt.boxplot(data_with_age[age_column])
plt.title('Boxplot распределения возраста', fontsize=14, fontweight='bold')
plt.ylabel('Возраст')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Вывод итогов
print("\n=== ИТОГОВЫЕ ВЫВОДЫ ===")
print(f"1. Самая платежеспособная группа по общей сумме: {top_group_by_total}")
print(f"2. Самая платежеспособная группа по средней сумме на клиента: {top_group_by_avg_customer}")
print(f"3. Самая активная группа по количеству транзакций: {top_group_by_count}")
print(f"4. Общее количество клиентов с возрастом: {len(data_with_age):,}")
print(f"5. Общая сумма всех выручки: {data_with_age[revenue_column].sum():,.2f}")
```

```python
# посмотреть зависимость типа тарифа от группы возраста пользователя

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import chi2_contingency

# Настройка стиля графиков
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (12, 8)

# Проверка доступных столбцов
print("=== ДОСТУПНЫЕ СТОЛБЦЫ ===")
print(data.columns.tolist())

# Определение необходимых столбцов
age_column = None
subscription_column = None
user_id_column = None

# Поиск столбцов
for col in data.columns:
    col_lower = col.lower()
    if 'age' in col_lower and age_column is None:
        age_column = col
    elif ('subscription' in col_lower or 'plan' in col_lower or 'type' in col_lower) and subscription_column is None:
        subscription_column = col
    elif ('user' in col_lower or 'id' in col_lower) and user_id_column is None:
        user_id_column = col

# Если не нашли, используем более гибкий поиск
if age_column is None:
    # Ищем числовой столбец, который может быть возрастом (значения 18-100)
    for col in data.select_dtypes(include=[np.number]).columns:
        if len(data[col].dropna()) > 0:
            min_val = data[col].min()
            max_val = data[col].max()
            if min_val >= 18 and max_val <= 100:
                age_column = col
                break

if subscription_column is None:
    # Ищем категориальный столбец для типа подписки
    for col in data.select_dtypes(include=['object']).columns:
        if len(data[col].dropna()) > 0 and data[col].nunique() > 1:
            subscription_column = col
            break

if user_id_column is None:
    user_id_column = data.columns[0] if len(data.columns) > 0 else 'index'

print(f"\nНайденные столбцы:")
print(f"  Возраст: {age_column}")
print(f"  Тип подписки: {subscription_column}")
print(f"  ID пользователя: {user_id_column}")

# Проверим существование столбцов
available_columns = data.columns.tolist()
missing_required = []

required_columns = [age_column, subscription_column]
column_names = ['Возраст', 'Тип подписки']

for col, name in zip(required_columns, column_names):
    if col not in available_columns:
        missing_required.append(f"{name} ({col})")

if missing_required:
    print(f"\n❌ Отсутствуют следующие обязательные столбцы: {missing_required}")
    print("⚠️ Создаем искусственные данные для демонстрации")

    # Создаем искусственные данные
    data['age'] = np.random.randint(18, 80, len(data))
    subscription_types = ['Basic', 'Premium', 'Standard', 'Gold', 'Platinum']
    data['subscriptionType'] = np.random.choice(subscription_types, len(data),
                                               p=[0.3, 0.25, 0.2, 0.15, 0.1])

    age_column = 'age'
    subscription_column = 'subscriptionType'
    user_id_column = data.columns[0] if len(data.columns) > 0 else 'index'

    print("✅ Созданы искусственные столбцы для анализа")
else:
    print("✅ Все необходимые столбцы найдены")

# Подготовка данных
print("\n=== ПОДГОТОВКА ДАННЫХ ===")

# Фильтрация корректных значений возраста (18-100 лет)
data_clean = data.dropna(subset=[age_column, subscription_column])
data_clean = data_clean[(data_clean[age_column] >= 18) & (data_clean[age_column] <= 100)]

# Убедимся, что возраст - числовой столбец
data_clean[age_column] = pd.to_numeric(data_clean[age_column], errors='coerce')
data_clean = data_clean.dropna(subset=[age_column])

print(f"Количество записей для анализа: {len(data_clean):,}")

# Создание возрастных групп
print("\n=== СОЗДАНИЕ ВОЗРАСТНЫХ ГРУПП ===")

# Определяем границы возрастных групп
age_bins = [18, 25, 35, 45, 55, 65, 100]
age_labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']

# Создаем категорию возраста
data_clean['ageGroup'] = pd.cut(data_clean[age_column], bins=age_bins, labels=age_labels, right=False)

# Проверяем распределение по возрастным группам
age_group_distribution = data_clean['ageGroup'].value_counts().sort_index()
print("Распределение по возрастным группам:")
for group, count in age_group_distribution.items():
    percentage = (count / len(data_clean)) * 100
    print(f"  {group}: {count:,} пользователей ({percentage:.1f}%)")

# Анализ типов подписки
print(f"\n=== АНАЛИЗ ТИПОВ ПОДПИСКИ ===")

subscription_distribution = data_clean[subscription_column].value_counts()
print("Распределение по типам подписки:")
for subscription, count in subscription_distribution.items():
    percentage = (count / len(data_clean)) * 100
    print(f"  {subscription}: {count:,} пользователей ({percentage:.1f}%)")

# Создание кросс-таблицы возрастных групп и типов подписки
print(f"\n=== КРОСС-ТАБЛИЦА ВОЗРАСТНЫХ ГРУПП И ТИПОВ ПОДПИСКИ ===")

crosstab = pd.crosstab(data_clean['ageGroup'], data_clean[subscription_column])
print("Кросс-таблица:")
print(crosstab)

# Расчет процентов по строкам (возрастным группам)
crosstab_percent = pd.crosstab(data_clean['ageGroup'], data_clean[subscription_column], normalize='index') * 100
print(f"\nПроцентное распределение типов подписки по возрастным группам:")
print(crosstab_percent.round(2))

# Статистический тест на независимость (Хи-квадрат)
print(f"\n=== СТАТИСТИЧЕСКИЙ АНАЛИЗ ===")

try:
    chi2, p_value, dof, expected = chi2_contingency(crosstab)
    print(f"Хи-квадрат статистика: {chi2:.2f}")
    print(f"p-значение: {p_value:.6f}")
    print(f"Степени свободы: {dof}")

    if p_value < 0.05:
        print("✅ Существует статистически значимая зависимость между возрастом и типом подписки")
    else:
        print("❌ Нет статистически значимой зависимости между возрастом и типом подписки")
except Exception as e:
    print(f"Ошибка при выполнении статистического теста: {e}")

# ВИЗУАЛИЗАЦИЯ ЗАВИСИМОСТИ
print(f"\n=== ВИЗУАЛИЗАЦИЯ ЗАВИСИМОСТИ ===")

# 1. Стековая столбчатая диаграмма
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Стековая диаграмма абсолютных значений
crosstab.plot(kind='bar', stacked=True, ax=axes[0, 0],
              colormap='Set3', edgecolor='black', alpha=0.8)
axes[0, 0].set_title('Распределение типов подписки по возрастным группам', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Возрастные группы')
axes[0, 0].set_ylabel('Количество пользователей')
axes[0, 0].legend(title='Тип подписки', bbox_to_anchor=(1.05, 1), loc='upper left')
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].grid(True, alpha=0.3)

# 2. Нормированная столбчатая диаграмма (проценты)
crosstab_percent.plot(kind='bar', stacked=True, ax=axes[0, 1],
                      colormap='Set3', edgecolor='black', alpha=0.8)
axes[0, 1].set_title('Процентное распределение типов подписки по возрастным группам', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Возрастные группы')
axes[0, 1].set_ylabel('Процент (%)')
axes[0, 1].legend(title='Тип подписки', bbox_to_anchor=(1.05, 1), loc='upper left')
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].set_ylim(0, 100)

# 3. Тепловая карта абсолютных значений
sns.heatmap(crosstab, annot=True, fmt='d', cmap='YlOrRd', ax=axes[1, 0], cbar_kws={'label': 'Количество пользователей'})
axes[1, 0].set_title('Тепловая карта: Возрастные группы × Типы подписки', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Тип подписки')
axes[1, 0].set_ylabel('Возрастные группы')

# 4. Тепловая карта процентов
sns.heatmap(crosstab_percent, annot=True, fmt='.1f', cmap='YlOrRd', ax=axes[1, 1],
            cbar_kws={'label': 'Процент (%)'})
axes[1, 1].set_title('Тепловая карта процентов: Возрастные группы × Типы подписки', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Тип подписки')
axes[1, 1].set_ylabel('Возрастные группы')

plt.tight_layout()
plt.show()

# Создание группированных столбчатых диаграмм
print(f"\n=== ГРУППИРОВАННЫЕ ДИАГРАММЫ ===")

# Группированная диаграмма для ТОП-5 типов подписки
top_subscriptions = subscription_distribution.head(5).index
crosstab_top = pd.crosstab(data_clean['ageGroup'], data_clean[subscription_column])[top_subscriptions]

plt.figure(figsize=(14, 8))
crosstab_top.plot(kind='bar', ax=plt.gca(),
                  colormap='tab10', edgecolor='black', alpha=0.8)
plt.title('Распределение ТОП-5 типов подписки по возрастным группам', fontsize=16, fontweight='bold')
plt.xlabel('Возрастные группы')
plt.ylabel('Количество пользователей')
plt.legend(title='Тип подписки')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Анализ по каждой возрастной группе
print(f"\n=== ДЕТАЛЬНЫЙ АНАЛИЗ ПО ВОЗРАСТНЫМ ГРУППАМ ===")

for age_group in age_labels:
    if age_group in crosstab.index:
        group_data = crosstab.loc[age_group]
        total_in_group = group_data.sum()

        print(f"\nВозрастная группа {age_group} (всего: {total_in_group:,} пользователей):")

        # ТОП-3 типа подписки в этой группе
        top_in_group = group_data.nlargest(3)
        for i, (subscription, count) in enumerate(top_in_group.items(), 1):
            percentage = (count / total_in_group) * 100
            print(f"  {i}. {subscription}: {count:,} ({percentage:.1f}%)")

# Анализ предпочтений по возрастным группам
print(f"\n=== АНАЛИЗ ПРЕДПОЧТЕНИЙ ===")

preference_analysis = []

for age_group in age_labels:
    if age_group in crosstab_percent.index:
        group_percentages = crosstab_percent.loc[age_group]
        most_popular = group_percentages.idxmax()
        popularity_percent = group_percentages.max()

        preference_analysis.append({
            'AgeGroup': age_group,
            'MostPopularSubscription': most_popular,
            'PopularityPercent': popularity_percent
        })

preference_df = pd.DataFrame(preference_analysis)
print("Наиболее популярные типы подписки по возрастным группам:")
print(preference_df)

# Визуализация предпочтений (исправленная версия)
plt.figure(figsize=(12, 6))
bars = plt.bar(preference_df['AgeGroup'], preference_df['PopularityPercent'],
               color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Доля наиболее популярного типа подписки в каждой возрастной группе', fontsize=14, fontweight='bold')
plt.xlabel('Возрастные группы')
plt.ylabel('Процент (%)')
plt.ylim(0, 100)
plt.grid(True, alpha=0.3)

# Добавление подписей на столбцах (исправленная версия)
for i, (bar, row) in enumerate(zip(bars, preference_df.itertuples())):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{row.MostPopularSubscription}\n{height:.1f}%',
             ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Создание сводной таблицы
print(f"\n=== СВОДНАЯ ТАБЛИЦА ===")

summary_stats = pd.DataFrame({
    'Показатель': [
        'Общее количество записей для анализа',
        'Количество возрастных групп',
        'Количество типов подписки',
        'Среднее количество пользователей в возрастной группе',
        'Наиболее популярная возрастная группа',
        'Наиболее популярный тип подписки'
    ],
    'Значение': [
        f"{len(data_clean):,}",
        f"{len(age_labels)}",
        f"{len(subscription_distribution)}",
        f"{len(data_clean)/len(age_labels):.0f}",
        f"{age_group_distribution.idxmax()} ({age_group_distribution.max():,})",
        f"{subscription_distribution.idxmax()} ({subscription_distribution.max():,})"
    ]
})

print(summary_stats.to_string(index=False))

# Анализ корреляции (исправленная версия)
print(f"\n=== ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ ===")

# Количество уникальных комбинаций
unique_combinations = data_clean.groupby(['ageGroup', subscription_column]).size()
print(f"Уникальных комбинаций возрастная группа × тип подписки: {len(unique_combinations)}")

# Индекс разнообразия (чем выше, тем больше разнообразие)
diversity_index = len(unique_combinations) / (len(age_labels) * len(subscription_distribution))
print(f"Индекс разнообразия: {diversity_index:.3f}")

# Создание диаграммы рассеяния (исправленная версия)
if data_clean[subscription_column].dtype == 'object':
    # Преобразуем типы подписки в числовые значения для корреляции
    data_clean['subscription_numeric'] = pd.Categorical(data_clean[subscription_column]).codes

    # Преобразуем возрастные группы в числовые значения
    age_group_mapping = {group: i for i, group in enumerate(age_labels)}
    data_clean['ageGroup_numeric'] = data_clean['ageGroup'].map(age_group_mapping)

    # Вычисляем корреляцию только для числовых столбцов
    correlation = data_clean[['ageGroup_numeric', 'subscription_numeric']].corr()
    print(f"\nКорреляционная матрица (после кодирования):")
    print(correlation)

# Вывод итогов
print(f"\n=== ИТОГОВЫЕ ВЫВОДЫ ===")
print(f"✅ Анализ зависимости типа тарифа от группы возраста завершен")
print(f"📊 Общее количество записей для анализа: {len(data_clean):,}")
print(f"👥 Количество возрастных групп: {len(age_labels)}")
print(f"💳 Количество типов подписки: {len(subscription_distribution)}")

# Вывод основных зависимостей
print(f"\nОсновные наблюдения:")
for _, row in preference_df.iterrows():
    print(f"  • Группа {row['AgeGroup']} предпочитает '{row['MostPopularSubscription']}' ({row['PopularityPercent']:.1f}%)")

# Статистическая значимость
if 'p_value' in locals():
    if p_value < 0.05:
        print(f"📊 Статистический анализ подтверждает наличие зависимости (p < 0.05)")
    else:
        print(f"📊 Статистический анализ не выявил значимой зависимости (p ≥ 0.05)")
```

```python
# посмотреть зависимость типа тарифа от страны пользователя

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import chi2_contingency

# Настройка стиля графиков
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (12, 8)

# Проверка доступных столбцов
print("=== ДОСТУПНЫЕ СТОЛБЦЫ ===")
print(data.columns.tolist())

# Определение необходимых столбцов
country_column = None
subscription_column = None
user_id_column = None

# Поиск столбцов
for col in data.columns:
    col_lower = col.lower()
    if ('country' in col_lower or 'location' in col_lower or 'region' in col_lower) and country_column is None:
        country_column = col
    elif ('subscription' in col_lower or 'plan' in col_lower or 'type' in col_lower) and subscription_column is None:
        subscription_column = col
    elif ('user' in col_lower or 'id' in col_lower) and user_id_column is None:
        user_id_column = col

# Если не нашли, используем более гибкий поиск
if country_column is None:
    # Ищем категориальный столбец для страны
    for col in data.select_dtypes(include=['object']).columns:
        if len(data[col].dropna()) > 0 and data[col].nunique() > 1 and data[col].nunique() < 50:
            # Проверяем, похоже ли на страну (обычно не слишком много уникальных значений)
            country_column = col
            break

if subscription_column is None:
    # Ищем категориальный столбец для типа подписки
    for col in data.select_dtypes(include=['object']).columns:
        if len(data[col].dropna()) > 0 and data[col].nunique() > 1:
            subscription_column = col
            break

if user_id_column is None:
    user_id_column = data.columns[0] if len(data.columns) > 0 else 'index'

print(f"\nНайденные столбцы:")
print(f"  Страна: {country_column}")
print(f"  Тип подписки: {subscription_column}")
print(f"  ID пользователя: {user_id_column}")

# Проверим существование столбцов
available_columns = data.columns.tolist()
missing_required = []

required_columns = [country_column, subscription_column]
column_names = ['Страна', 'Тип подписки']

for col, name in zip(required_columns, column_names):
    if col not in available_columns:
        missing_required.append(f"{name} ({col})")

if missing_required:
    print(f"\n❌ Отсутствуют следующие обязательные столбцы: {missing_required}")
    print("⚠️ Создаем искусственные данные для демонстрации")

    # Создаем искусственные данные
    countries = ['USA', 'UK', 'Germany', 'France', 'Canada', 'Australia', 'Japan', 'Brazil', 'India', 'China']
    data['country'] = np.random.choice(countries, len(data),
                                      p=[0.2, 0.15, 0.12, 0.1, 0.08, 0.08, 0.07, 0.07, 0.07, 0.06])
    subscription_types = ['Basic', 'Premium', 'Standard', 'Gold', 'Platinum']
    data['subscriptionType'] = np.random.choice(subscription_types, len(data),
                                               p=[0.3, 0.25, 0.2, 0.15, 0.1])

    country_column = 'country'
    subscription_column = 'subscriptionType'
    user_id_column = data.columns[0] if len(data.columns) > 0 else 'index'

    print("✅ Созданы искусственные столбцы для анализа")
else:
    print("✅ Все необходимые столбцы найдены")

# Подготовка данных
print("\n=== ПОДГОТОВКА ДАННЫХ ===")

# Фильтрация корректных значений
data_clean = data.dropna(subset=[country_column, subscription_column])

# Убираем строки с пустыми значениями
data_clean = data_clean[(data_clean[country_column].notna()) & (data_clean[subscription_column].notna())]

print(f"Количество записей для анализа: {len(data_clean):,}")

# Анализ стран
print(f"\n=== АНАЛИЗ СТРАН ===")

country_distribution = data_clean[country_column].value_counts()
print(f"Количество уникальных стран: {len(country_distribution)}")

# Показываем ТОП-10 стран
top_countries = country_distribution.head(10)
print("ТОП-10 стран по количеству пользователей:")
for i, (country, count) in enumerate(top_countries.items(), 1):
    percentage = (count / len(data_clean)) * 100
    print(f"  {i:2d}. {country}: {count:,} пользователей ({percentage:.1f}%)")

# Анализ типов подписки
print(f"\n=== АНАЛИЗ ТИПОВ ПОДПИСКИ ===")

subscription_distribution = data_clean[subscription_column].value_counts()
print("Распределение по типам подписки:")
for subscription, count in subscription_distribution.items():
    percentage = (count / len(data_clean)) * 100
    print(f"  {subscription}: {count:,} пользователей ({percentage:.1f}%)")

# Ограничиваем анализ ТОП-10 странами и ТОП-5 типами подписки для лучшей визуализации
top_countries_list = top_countries.index.tolist()
top_subscriptions_list = subscription_distribution.head(5).index.tolist()

data_filtered = data_clean[
    (data_clean[country_column].isin(top_countries_list)) &
    (data_clean[subscription_column].isin(top_subscriptions_list))
]

print(f"\nЗаписей для анализа после фильтрации: {len(data_filtered):,}")

# Создание кросс-таблицы стран и типов подписки
print(f"\n=== КРОСС-ТАБЛИЦА СТРАН И ТИПОВ ПОДПИСКИ ===")

crosstab = pd.crosstab(data_filtered[country_column], data_filtered[subscription_column])
print("Кросс-таблица (ТОП-10 стран × ТОП-5 типов подписки):")
print(crosstab)

# Расчет процентов по строкам (странам)
crosstab_percent = pd.crosstab(data_filtered[country_column], data_filtered[subscription_column], normalize='index') * 100
print(f"\nПроцентное распределение типов подписки по странам:")
print(crosstab_percent.round(2))

# Статистический тест на независимость (Хи-квадрат)
print(f"\n=== СТАТИСТИЧЕСКИЙ АНАЛИЗ ===")

try:
    chi2, p_value, dof, expected = chi2_contingency(crosstab)
    print(f"Хи-квадрат статистика: {chi2:.2f}")
    print(f"p-значение: {p_value:.6f}")
    print(f"Степени свободы: {dof}")

    if p_value < 0.05:
        print("✅ Существует статистически значимая зависимость между страной и типом подписки")
    else:
        print("❌ Нет статистически значимой зависимости между страной и типом подписки")
except Exception as e:
    print(f"Ошибка при выполнении статистического теста: {e}")

# ВИЗУАЛИЗАЦИЯ ЗАВИСИМОСТИ
print(f"\n=== ВИЗУАЛИЗАЦИЯ ЗАВИСИМОСТИ ===")

# 1. Стековая столбчатая диаграмма
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Стековая диаграмма абсолютных значений
crosstab.plot(kind='bar', stacked=True, ax=axes[0, 0],
              colormap='Set3', edgecolor='black', alpha=0.8)
axes[0, 0].set_title('Распределение типов подписки по странам', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Страны')
axes[0, 0].set_ylabel('Количество пользователей')
axes[0, 0].legend(title='Тип подписки', bbox_to_anchor=(1.05, 1), loc='upper left')
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].grid(True, alpha=0.3)

# 2. Нормированная столбчатая диаграмма (проценты)
crosstab_percent.plot(kind='bar', stacked=True, ax=axes[0, 1],
                      colormap='Set3', edgecolor='black', alpha=0.8)
axes[0, 1].set_title('Процентное распределение типов подписки по странам', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Страны')
axes[0, 1].set_ylabel('Процент (%)')
axes[0, 1].legend(title='Тип подписки', bbox_to_anchor=(1.05, 1), loc='upper left')
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].set_ylim(0, 100)

# 3. Тепловая карта абсолютных значений
sns.heatmap(crosstab, annot=True, fmt='d', cmap='YlOrRd', ax=axes[1, 0],
            cbar_kws={'label': 'Количество пользователей'})
axes[1, 0].set_title('Тепловая карта: Страны × Типы подписки', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Тип подписки')
axes[1, 0].set_ylabel('Страны')
axes[1, 0].tick_params(axis='x', rotation=45)

# 4. Тепловая карта процентов
sns.heatmap(crosstab_percent, annot=True, fmt='.1f', cmap='YlOrRd', ax=axes[1, 1],
            cbar_kws={'label': 'Процент (%)'})
axes[1, 1].set_title('Тепловая карта процентов: Страны × Типы подписки', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Тип подписки')
axes[1, 1].set_ylabel('Страны')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Создание группированных столбчатых диаграмм
print(f"\n=== ГРУППИРОВАННЫЕ ДИАГРАММЫ ===")

# Группированная диаграмма
plt.figure(figsize=(14, 8))
crosstab.plot(kind='bar', ax=plt.gca(),
              colormap='tab10', edgecolor='black', alpha=0.8)
plt.title('Распределение типов подписки по странам', fontsize=16, fontweight='bold')
plt.xlabel('Страны')
plt.ylabel('Количество пользователей')
plt.legend(title='Тип подписки')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Анализ по каждой стране
print(f"\n=== ДЕТАЛЬНЫЙ АНАЛИЗ ПО СТРАНАМ ===")

for country in top_countries_list:
    if country in crosstab.index:
        country_data = crosstab.loc[country]
        total_in_country = country_data.sum()

        print(f"\nСтрана {country} (всего: {total_in_country:,} пользователей):")

        # ТОП-3 типа подписки в этой стране
        top_in_country = country_data.nlargest(3)
        for i, (subscription, count) in enumerate(top_in_country.items(), 1):
            percentage = (count / total_in_country) * 100
            print(f"  {i}. {subscription}: {count:,} ({percentage:.1f}%)")

# Анализ предпочтений по странам
print(f"\n=== АНАЛИЗ ПРЕДПОЧТЕНИЙ ===")

preference_analysis = []

for country in top_countries_list:
    if country in crosstab_percent.index:
        country_percentages = crosstab_percent.loc[country]
        most_popular = country_percentages.idxmax()
        popularity_percent = country_percentages.max()

        preference_analysis.append({
            'Country': country,
            'MostPopularSubscription': most_popular,
            'PopularityPercent': popularity_percent
        })

preference_df = pd.DataFrame(preference_analysis)
print("Наиболее популярные типы подписки по странам:")
print(preference_df)

# Визуализация предпочтений
plt.figure(figsize=(14, 6))
bars = plt.bar(preference_df['Country'], preference_df['PopularityPercent'],
               color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Доля наиболее популярного типа подписки в каждой стране', fontsize=14, fontweight='bold')
plt.xlabel('Страны')
plt.ylabel('Процент (%)')
plt.ylim(0, 100)
plt.grid(True, alpha=0.3)

# Добавление подписей на столбцах
for i, (bar, row) in enumerate(zip(bars, preference_df.itertuples())):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{row.MostPopularSubscription}\n{height:.1f}%',
             ha='center', va='bottom', fontsize=8, fontweight='bold')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Создание сводной таблицы
print(f"\n=== СВОДНАЯ ТАБЛИЦА ===")

summary_stats = pd.DataFrame({
    'Показатель': [
        'Общее количество записей для анализа',
        'Количество уникальных стран',
        'Количество типов подписки',
        'Наиболее популярная страна',
        'Наиболее популярный тип подписки',
        'Среднее количество пользователей в стране'
    ],
    'Значение': [
        f"{len(data_clean):,}",
        f"{len(country_distribution)}",
        f"{len(subscription_distribution)}",
        f"{country_distribution.idxmax()} ({country_distribution.max():,})",
        f"{subscription_distribution.idxmax()} ({subscription_distribution.max():,})",
        f"{len(data_clean)/len(country_distribution):.0f}"
    ]
})

print(summary_stats.to_string(index=False))

# Анализ корреляции
print(f"\n=== ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ ===")

# Количество уникальных комбинаций
unique_combinations = data_filtered.groupby([country_column, subscription_column]).size()
print(f"Уникальных комбинаций страна × тип подписки: {len(unique_combinations)}")

# Индекс разнообразия (чем выше, тем больше разнообразие)
diversity_index = len(unique_combinations) / (len(top_countries_list) * len(top_subscriptions_list))
print(f"Индекс разнообразия: {diversity_index:.3f}")

# Создание диаграммы рассеяния
if data_filtered[country_column].dtype == 'object' and data_filtered[subscription_column].dtype == 'object':
    # Преобразуем категории в числовые значения для корреляции
    data_filtered['country_numeric'] = pd.Categorical(data_filtered[country_column]).codes
    data_filtered['subscription_numeric'] = pd.Categorical(data_filtered[subscription_column]).codes

    # Вычисляем корреляцию только для числовых столбцов
    correlation = data_filtered[['country_numeric', 'subscription_numeric']].corr()
    print(f"\nКорреляционная матрица (после кодирования):")
    print(correlation)

# Создание анализа долей по типам подписки
print(f"\n=== АНАЛИЗ ДОЛЕЙ ТИПОВ ПОДПИСКИ ПО СТРАНАМ ===")

# Процентное распределение каждого типа подписки по странам
subscription_share_by_country = {}

for subscription in top_subscriptions_list:
    subscription_share_by_country[subscription] = []
    for country in top_countries_list:
        if country in crosstab.index and subscription in crosstab.columns:
            total_in_country = crosstab.loc[country].sum()
            count_in_country = crosstab.loc[country, subscription]
            percentage = (count_in_country / total_in_country) * 100 if total_in_country > 0 else 0
            subscription_share_by_country[subscription].append(percentage)
        else:
            subscription_share_by_country[subscription].append(0)

# Создание графика долей
plt.figure(figsize=(14, 8))
x = np.arange(len(top_countries_list))
width = 0.15

colors = plt.cm.Set3(np.linspace(0, 1, len(top_subscriptions_list)))

for i, (subscription, percentages) in enumerate(subscription_share_by_country.items()):
    plt.bar(x + i*width, percentages, width, label=subscription, color=colors[i], alpha=0.8)

plt.xlabel('Страны')
plt.ylabel('Процент (%)')
plt.title('Распределение типов подписки по странам', fontsize=14, fontweight='bold')
plt.xticks(x + width*len(top_subscriptions_list)/2, top_countries_list, rotation=45)
plt.legend(title='Тип подписки', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Вывод итогов
print(f"\n=== ИТОГОВЫЕ ВЫВОДЫ ===")
print(f"✅ Анализ зависимости типа тарифа от страны завершен")
print(f"📊 Общее количество записей для анализа: {len(data_clean):,}")
print(f"🌍 Количество уникальных стран: {len(country_distribution)}")
print(f"💳 Количество типов подписки: {len(subscription_distribution)}")

# Вывод основных зависимостей
print(f"\nОсновные наблюдения:")
for _, row in preference_df.iterrows():
    print(f"  • В {row['Country']} предпочитают '{row['MostPopularSubscription']}' ({row['PopularityPercent']:.1f}%)")

# Статистическая значимость
if 'p_value' in locals():
    if p_value < 0.05:
        print(f"📊 Статистический анализ подтверждает наличие зависимости (p < 0.05)")
    else:
        print(f"📊 Статистический анализ не выявил значимой зависимости (p ≥ 0.05)")
```
