# Приведение столбцов к стилю camel_case

```python
import pandas as pd

# Пример: если у тебя уже есть DataFrame data
# data = pd.read_csv('your_file.csv')  # или откуда-то ещё

# Функция для преобразования названий столбцов в camelCase
def to_camel_case(snake_str):
    components = snake_str.replace('(', '').replace(')', '').split('_')
    return components[0] + ''.join(x.capitalize() for x in components[1:])

# Приведение названий столбцов к camelCase
data.columns = [to_camel_case(col) for col in data.columns]

# Посмотреть новые названия столбцов
print(data.columns.tolist())
```

```python
# изучить распределение продаж по категориям заказов

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # если хотите использовать палитру seaborn

# Группировка данных
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
sales_by_category = sales_by_category.sort_values(by='Sales', ascending=False)
total_sales = sales_by_category['Sales'].sum()

# Функция для подписей абсолют/процент
def make_autopct(values):
    def inner_autopct(pct):
        total = sum(values)
        absolute = int(round(pct*total/100.0))
        return "{:.1f}%\n({:,.0f})".format(pct, absolute)
    return inner_autopct

# Получаем столько цветов, сколько уникальных категорий:
colors = sns.color_palette('Set2', n_colors=len(sales_by_category['Category']))
# Альтернативно без seaborn:
# import numpy as np
# colors = plt.cm.Pastel1(np.arange(len(sales_by_category['Category'])))

plt.figure(figsize=(8, 8))
plt.pie(
    sales_by_category['Sales'],
    labels=sales_by_category['Category'],
    autopct=make_autopct(sales_by_category['Sales']),
    startangle=90,
    colors=colors
)
plt.title('Доли продаж по категориям заказов')
plt.axis('equal')
plt.tight_layout()
plt.show()

```

```python
# изучить распределение продаж по городам (штатам)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Группируем продажи по городам
sales_by_city = data.groupby('City')['Sales'].sum().reset_index()

# Указываем количество топ-города, которые хотим отобразить отдельно
top_n = 10

# Отбираем топ-N городов по продажам
sales_top_cities = sales_by_city.sort_values(by='Sales', ascending=False).head(top_n)

# Суммируем продажи для всех остальных городов
other_sales = sales_by_city.sort_values(by='Sales', ascending=False).iloc[top_n:]['Sales'].sum()

# Создаем DataFrame для категории "Другие"
other_df = pd.DataFrame([{'City': 'Другие', 'Sales': other_sales}])

# Объединяем топовые города и "Другие" в один DataFrame
sales_top_cities = pd.concat([sales_top_cities, other_df], ignore_index=True)

# Функция для подписей с абсолютными и процентными значениями
def make_autopct(values):
    total = sum(values)
    # Возвращаем список подписей в формате "значение (процент)"
    return ["{:,.0f} ({:.1f}%)".format(v, v / total * 100) for v in values]

# Получаем подписи для столбцов
labels = make_autopct(sales_top_cities['Sales'])

# Цветовая палитра для разноцветной диаграммы
colors = sns.color_palette('tab10', n_colors=len(sales_top_cities))

plt.figure(figsize=(10, 6))
bars = plt.bar(
    sales_top_cities['City'],
    sales_top_cities['Sales'],
    color=colors
)

# Добавляем подписи над столбцами
for bar, label in zip(bars, labels):
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        label,
        ha='center',
        va='bottom',
        fontsize=9
    )

plt.title('Распределение продаж по городам (топ-10 и другие)')
plt.ylabel('Сумма продаж')
plt.xlabel('Город')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
# самые популярные товары (Product Name) по сумме продаж

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Группируем данные по названию товара и суммируем продажи
sales_by_product = data.groupby('Product Name')['Sales'].sum().reset_index()

# Сортируем по убыванию продаж и выбираем топ-10 товаров
top_n = 10
top_products = sales_by_product.sort_values(by='Sales', ascending=False).head(top_n)

# Общая сумма продаж для вычисления процентов
total_sales = top_products['Sales'].sum()

# Цветовая палитра для гистограммы
colors = sns.color_palette('tab10', n_colors=top_n)

plt.figure(figsize=(12, 6))
bars = plt.bar(top_products['Product Name'], top_products['Sales'], color=colors)

# Добавляем подписи с абсолютными и процентными значениями продаж над столбцами
for bar in bars:
    height = bar.get_height()
    percent = (height / total_sales) * 100
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height:,.0f}\n({percent:.1f}%)',
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.title('Топ-10 самых популярных товаров по сумме продаж')
plt.ylabel('Сумма продаж')
plt.xlabel('Наименование товара')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
# самые популярные товары (Product Name) по количеству продаж

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Группируем данные по названию товара и суммируем количество продаж
quantity_by_product = data.groupby('Product Name')['Quantity'].sum().reset_index()

# Сортируем по убыванию количества и выбираем топ-10 товаров
top_n = 10
top_products = quantity_by_product.sort_values(by='Quantity', ascending=False).head(top_n)

# Общая сумма продаж для вычисления процентов
total_quantity = top_products['Quantity'].sum()

# Цветовая палитра для гистограммы
colors = sns.color_palette('tab10', n_colors=top_n)

plt.figure(figsize=(12, 6))
bars = plt.bar(top_products['Product Name'], top_products['Quantity'], color=colors)

# Добавляем подписи с абсолютными и процентными значениями количества продаж над столбцами
for bar in bars:
    height = bar.get_height()
    percent = (height / total_quantity) * 100
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{int(height):,}\n({percent:.1f}%)',
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.title('Топ-10 самых популярных товаров по количеству продаж')
plt.ylabel('Количество продаж')
plt.xlabel('Наименование товара')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
# изучить распределение категорий внутри городов (штатам)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Группируем данные по городу (или штату) и категории товара, суммируем продажи
# Для анализа по городу:
grouped = data.groupby(['City', 'Category'])['Sales'].sum().reset_index()

# Если хотите по штату, замените 'City' на 'State' в группировке:
# grouped = data.groupby(['State', 'Category'])['Sales'].sum().reset_index()

# Создаём сводную таблицу: строки — города, столбцы — категории, значения — суммы продаж
pivot = grouped.pivot(index='City', columns='Category', values='Sales').fillna(0)

# Выбираем топ-N городов по общей сумме продаж, чтобы диаграмма была читаемой
top_n = 10
top_cities = pivot.sum(axis=1).sort_values(ascending=False).head(top_n).index
pivot_top = pivot.loc[top_cities]

# Цветовая палитра для категорий товаров
colors = sns.color_palette('Set2', n_colors=len(pivot_top.columns))

# Создаём фигуру и оси
fig, ax = plt.subplots(figsize=(12, 7))

# Начальное значение для накопления высоты баров
bottom = np.zeros(len(pivot_top))

# Для расчёта процентов по городам (строкам)
row_sums = pivot_top.sum(axis=1).values

# Построение накопленных столбцов с подписями абсолютных и относительных значений
for i, category in enumerate(pivot_top.columns):
    values = pivot_top[category].values
    bars = ax.bar(pivot_top.index, values, bottom=bottom, color=colors[i], label=category)
    for bar, val, total in zip(bars, values, row_sums):
        if val > 0:
            height = bar.get_height()
            percent = (val / total) * 100
            ax.text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_y() + height / 2,
                f'{int(val):,}\n({percent:.1f}%)',
                ha='center',
                va='center',
                fontsize=9,
                color='black'
            )
    bottom += values  # Обновляем нижнюю границу для следующего сегмента

ax.set_title('Распределение продаж по категориям товаров в городах (топ-10) с накоплением и %')
ax.set_xlabel('Город')
ax.set_ylabel('Сумма продаж')
ax.legend(title='Категория товара', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# изучить динамику продаж внутри категорий

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Предположим, что 'Order Date' в формате datetime
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Создаем столбец с периодом (например, год-месяц)
data['YearMonth'] = data['Order Date'].dt.to_period('M').astype(str)

# Группируем по периоду и категории с суммированием продаж
sales_trend = data.groupby(['YearMonth', 'Category'])['Sales'].sum().reset_index()

# Для удобства построения используем сводную таблицу: строки — периоды, столбцы — категории
pivot_trend = sales_trend.pivot(index='YearMonth', columns='Category', values='Sales').fillna(0)

# Настроим размер графика и палитру цветов
plt.figure(figsize=(14, 7))
colors = sns.color_palette('Set2', n_colors=len(pivot_trend.columns))

# Построим линейные графики динамики по каждой категории
for i, category in enumerate(pivot_trend.columns):
    plt.plot(pivot_trend.index, pivot_trend[category], label=category, color=colors[i], marker='o')

plt.title('Динамика продаж по товарным категориям')
plt.xlabel('Период (год-месяц)')
plt.ylabel('Сумма продаж')
plt.xticks(rotation=45)
plt.legend(title='Категории товаров')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression

# Предположим, что 'Order Date' уже в datetime
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Создаём столбец с периодом (например, год-месяц)
data['YearMonth'] = data['Order Date'].dt.to_period('M').astype(str)

# Группируем продажи по периоду и категории
sales_trend = data.groupby(['YearMonth', 'Category'])['Sales'].sum().reset_index()

# Для столбчатой диаграммы — агрегация по периоду суммарно по всем категориям
sales_total = sales_trend.groupby('YearMonth')['Sales'].sum().reset_index()

# Для линий тренда — сводная таблица {YearMonth x Category}
pivot = sales_trend.pivot(index='YearMonth', columns='Category', values='Sales').fillna(0)

# Преобразуем индекс (YearMonth) в порядковый числовой формат для регрессии
x_vals = np.arange(len(pivot)).reshape(-1, 1)

# Создаём график
plt.figure(figsize=(14, 8))
colors = sns.color_palette('Set2', n_colors=len(pivot.columns))

# Столбчатая диаграмма: суммарные продажи по всем категориям на периоды
plt.bar(sales_total['YearMonth'], sales_total['Sales'], color='lightgray', alpha=0.7, label='Общие продажи')

# Линии тренда для каждой категории
for i, category in enumerate(pivot.columns):
    y_vals = pivot[category].values
    # Регрессия для линии тренда
    model = LinearRegression()
    model.fit(x_vals, y_vals)
    y_trend = model.predict(x_vals)

    # График линий тренда (с маркерами)
    plt.plot(pivot.index, y_trend, label=f'{category} (тренд)', color=colors[i], linewidth=2)
    # Также можно показать реальные данные по категориям как точки
    plt.scatter(pivot.index, y_vals, color=colors[i], alpha=0.6)

plt.title('Динамика продаж по периодам с линиями тренда по товарным категориям')
plt.xlabel('Период (год-месяц)')
plt.ylabel('Сумма продаж')
plt.xticks(rotation=45)
plt.legend(title='Категории и тренды', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
